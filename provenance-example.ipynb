```python
import requests

request="""{
  "provenance": {
    "request": {
      "maxResults": 10,
      "startDate": "06/24/2019 00:00:00 EST",
      "endDate": "06/25/2019 23:59:59 EST",
      "searchTerms": {
        "ProcessorID": "86271673-016b-1000-ffff-ffff9c33a3fe"
      }
    }
  }
}"""
headers={}
headers['content-Type']='application/json'
base_url="http://localhost:9090/nifi-api/provenance"
response = requests.post(base_url,data=request,headers=headers)
print(response.content)
```

```python
import json
responseObj=json.loads(response.content)
prov_url = responseObj["provenance"]["uri"]
print(prov_url)

response=requests.get(prov_url,headers)
```

```python
print(json.dumps(json.loads(response.content), indent=2))
```
