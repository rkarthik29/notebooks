```python
board=[["5","3",".",".","7",".",".",".","."],
       ["6",".",".","1","9","5",".",".","."],
       [".","9","8",".",".",".",".","6","."],
       ["8",".",".",".","6",".",".",".","3"],
       ["4",".",".","8",".","3",".",".","1"],
       ["7",".",".",".","2",".",".",".","6"],
       [".","6",".",".",".",".","2","8","."],
       [".",".",".","4","1","9",".",".","5"],
       [".",".",".",".","8",".",".","7","9"]]

def isValidSudoku(board):
    """
    :type board: List[List[str]]
    :rtype: bool
    """
    rows=[]
    cols=[]
    squares=[]
    for i in range(len(board)):
        rows.append(list())
        cols.append(list())
        squares.append(list())
    for i in range(len(board)):
        for j in range(len(board)):
            
            if(board[i][j]=='.'):
                continue
            print(board[i][j],i,j)
            if( board[i][j] in cols[j]):
                return False
            else:
                cols[j].append(board[i][j])
            if( board[i][j] in rows[i]):
                return False
            else:
                rows[i].append(board[i][j])
            sq_index=int(i/3)*3 +int(j/3)
            if(board[i][j] in squares[sq_index]):
                return False
            else:
                squares[sq_index].append(board[i][j])
            print (squares,sq_index)
    return True

isValidSudoku(board)
```
